```python
# Omsk
#load
df = pd.read_csv('omsk_3year.csv', sep=';', skiprows=6)
df.rename(columns={ df.columns[0]: "time" }, inplace = True)
df['time'] =  pd.to_datetime(df['time'], format='%d.%m.%Y %H:%M')
df = df.set_index('time')
df = df.iloc[::-1]
df = df.drop(columns=['P', 'Pa', 'ff10', 'ff3', 'Ch', 'Td', 'E', 'Tg', 'W1', 'W2', 'Cl', 'Nh', 'H', 'Cm', "E'", 'Unnamed: 29'])
df.tail(100)
```

Airport (METAR)

```python
# Airport METAR
dfa = pd.read_csv('airport.csv', sep=';', skiprows=6)
dfa.rename(columns={ dfa.columns[0]: "time" }, inplace = True)
dfa['time'] =  pd.to_datetime(dfa['time'], format='%d.%m.%Y %H:%M')
dfa = dfa.set_index('time')
dfa = dfa.iloc[::-1]
dfa.tail(200)
```

```python
# filter by time
# filter by time
dfa_filtered = dfa.loc[dfa.index.time == datetime.time(12,0)]
dfa_filtered.tail(50)
```

```python
# get year data from both
df_omsk = df_filtered[datetime.date(2019,5,2):datetime.date(2020,5,3)]
df_airport = dfa_filtered[datetime.date(2019,5,2):datetime.date(2020,5,2)]

df_compare = df_omsk[['T']].copy()
df_compare['T_a'] = df_airport['T']
df_compare['diff'] = abs(df_compare['T'] - df_compare['T_a'])
df_compare.sort_values(by=['diff'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(35,10)})
sns.set(style="darkgrid")
sns.lineplot(x=df_omsk_last_year.index, y='T', data=df_omsk_last_year)
```

```python
# heat map
nan_elems = df_omsk_last_year['T'].isnull()
df_omsk_last_year = df_omsk_last_year[~nan_elems]
data_ = [df_omsk_last_year.index, df_omsk_last_year['T']]
ax = sns.heatmap(data_)
ax = sns.heatmap(data_)
```

```python
by_month = (df_omsk_last_year.groupby(df_omsk_last_year.index.month)['T']
                              .agg(['mean', 'min', 'max'])
                              .reset_index())
by_month.info()
```

```python
ax = sns.barplot(x=by_month.time, y="min", data=by_month)
```
