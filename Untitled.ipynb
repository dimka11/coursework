```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 8]
y = [41.1,0.3581,2.41533,8.88788, 40.8202]
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='n ', ylabel='время выполнения',
       title='количество уравнений / время выполнения')
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='n ', ylabel='количество процессов',
       title='количество процессов /время выполнения')
ax.grid()
plt.show()
```
