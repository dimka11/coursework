```python
def download_html():
    folder_name = 'NewData/html' # 
    os.chdir("B:/Projects/CourseWork/coursework/" + folder_name)
    print(os.getcwd())

    #create folders
    for i in range(2020, 2021 + 1):
        try:
            os.mkdir(f'{i}')
        except:
            pass
        
    headers = {
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:45.0) Gecko/20100101 Firefox/45.0'
          }

    days = 31
    station = 28698 # 28791 - Шербакуль # 28598 - Саргатское # 28696 - Калачинск # 28698 - Омск
    day = 1

    for year in range(2020, 2021 + 1):
        for month in range(1, 12 + 1):
            url = f'http://www.ogimet.com/cgi-bin/gsynres?ind={station}&lang=en&decoded=yes&ndays={days}&ano={year}&mes={month}&day={day}/'
            r = requests.get(url, headers = headers)
            with open(f'{year}/{month}{year}.html', 'w', encoding='utf8') as output_file:
                output_file.write(r.text)
            print(f'month:{month} in year:{year} was downloaded')
            
    os.chdir("B:/Projects/CourseWork/coursework/NewData"
```

```python
# all files
def parse_html():
    folder_name = 'NewData/html' # 
#     os.chdir("B:/Projects/CourseWork/coursework/" + folder_name)
    os.chdir('B:/Projects/CourseWork/coursework/HTMLCSV/ogimet/data/')
    print(os.getcwd())
    
    m_df = pd.DataFrame()
    
    for year in range(1999, 2021 + 1):
        for month in range(1, 12 + 1):
            print(f'month start parsing {month} {year}')
            try:
                file = open(f'{year}/{month}{year}.html', 'rb')
            except:
                continue
            
            soup = BeautifulSoup(file, "lxml")
            soup = soup.select('body > table> tr:nth-child(2) > td:nth-child(2) > table:nth-child(2) > thead')[0]

            list_title = [item.text for item in soup.select('tr:nth-child(1)')[0].find_all('th')]
            list_title.insert(1, 'time')

            rows = []
            for row in range(len(soup.find_all('tr'))):
                if row not in [0,1]:
                    row_items_list = [parse_row(item) for item in soup.select(f'tr:nth-child({row})')[0].find_all('td')]
                    rows.append(row_items_list)
            df_ = pd.DataFrame(rows, columns=list_title)
            df_['Datetime'] = pd.to_datetime(df_['Date'] + ' ' + df_['time'])
            df_ = df_.set_index('Datetime')
            df_ = df_.drop(columns=['Date', 'time'])
            df_ = df_.reindex(index=df_.index[::-1])
            m_df = m_df.append(df_)
            m_df.to_csv(f'Omsk_weather_{year}_{month}.csv')
    
    os.chdir("B:/Projects/CourseWork/coursework/NewData")
    return m_df
    
```

```python
# Basic clean
def clean(m_df):
    # drop dublicates
    m_df = m_df.loc[~m_df.index.duplicated(keep='first')]
    
    # replace '-----' to nan
    m_df['T(C)'].replace('-----', np.nan, inplace=True)
    m_df['Td(C)'].replace('-----', np.nan, inplace=True)
    m_df['Tmax(C)'].replace('-----', np.nan, inplace=True)
    m_df['Tmin(C)'].replace('-----', np.nan, inplace=True)
    
    m_df['T(C)'] = m_df['T(C)'].astype(float)
    m_df['Td(C)'] = m_df['Td(C)'].astype(float)
    m_df['Tmax(C)'] = m_df['Tmax(C)'].astype(float)
    m_df['Tmin(C)'] = m_df['Tmin(C)'].astype(float)
    
    # Clear incorrect
    m_df['T(C)'].replace(-90.4, np.nan, inplace=True)
    
    # replace NaN to mean of 2 nb
    m_df['T(C)'] = m_df['T(C)'].where(m_df['T(C)'].values == 999, other=(m_df['T(C)'].fillna(method='ffill') + m_df['T(C)'].fillna(method='bfill'))/2)
    
    m_df = m_df.shift(periods=2)
    m_df = m_df.iloc[2:]
    
    return m_df

m_df = clean(m_df)
m_df.info()
```
