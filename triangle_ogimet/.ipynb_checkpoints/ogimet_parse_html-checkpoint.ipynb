```python
# all files

import pandas as pd
import datetime
import ntpath
from bs4 import BeautifulSoup

m_df = pd.DataFrame()

for year in range(1999, 2020 + 1):
    for month in range(1, 12 + 1):
        print('month start parsing ' + str(month) + ' ' + str(year))
        file = ''
        try:
            file = open(f'{year}/{month}{year}.html', 'rb')
        except:
            continue
        
        
        # reversed and append to dataframe
        # append to dataframe
        data = []
        col_names = ['date', 'time', 'T', 'Td', 'Tmax', 'Tmin', 'ddd', 'ff', 'prec', 'nt', 'nh', 'h']

        soup = BeautifulSoup(file, "lxml")
        soup = soup.select('body > table> tr:nth-child(2) > td:nth-child(2) > table:nth-child(2) > thead')[0]

        for i, tr in enumerate(reversed(soup.findAll("tr"))):
            if i == len(soup.findAll("tr")) - 1:
                continue
            try:
                date = str(tr.select('td:nth-child(1) > a')[0].string)
                time = str(tr.select('td:nth-child(2) > a')[0].string)
                T = str(tr.select('td:nth-child(3)')[0].string)
                Td = str(tr.select('td:nth-child(4)')[0].string)
                Tmax = str(tr.select('td:nth-child(5)')[0].string)
                Tmin = str(tr.select('td:nth-child(6)')[0].string)
                ddd = str(tr.select('td:nth-child(7)')[0].string)
                ff = str(tr.select('td:nth-child(8)')[0].string)
#                 prec = str(tr.select('td:nth-child(13)')[0].string)
#                 nt = str(tr.select('td:nth-child(14)')[0].string)
#                 nh = str(tr.select('td:nth-child(15)')[0].string)
#                 h = str(tr.select('td:nth-child(16)')[0].string)
            #     print(date, end = ' ')
            #     print(time, end = ' ')
            #     print(temp)
                data.append({'date': date, 'time': time, 'T': T, 'Td': Td, 'Tmax': Tmax, 'Tmin': Tmin, 'ddd': ddd, 'ff':ff})
            except:
                print(file)
                raise

        df = pd.DataFrame(data, columns = col_names)
        if m_df.empty:
            m_df = df
        else:
            m_df = m_df.append(df, ignore_index=True)

# m_df.drop_duplicates() # not work because not implace
m_df.to_csv('weather_data.csv')
```

```python
# load
import pandas as pd
import numpy as np

df = pd.read_csv('weather_data.csv') # dfao = pd.read_csv('airport_ogimet_ext.csv')
df['time1'] = df['date'] + ' ' + df['time']
df = df.drop(columns={'Unnamed: 0', 'date', 'time'})
df['time1'] =  pd.to_datetime(df['time1'], format='%m/%d/%Y %H:%M')
df.rename(columns={ 'time1': "time" }, inplace = True)
df = df.set_index('time')

# drop dublicates
df = df.loc[~df.index.duplicated(keep='first')]

# replace '-----' to nan
df['T'].replace('-----', np.nan, inplace=True)

# df['Tmax'].replace('-----', np.nan, inplace=True)
# df['Tmin'].replace('-----', np.nan, inplace=True)
# df['Td'].replace('-----', np.nan, inplace=True)

df['T'] = df['T'].astype(float)
# df['Tmax'] = df['Tmax'].astype(float)
# df['Tmin'] = df['Tmin'].astype(float)
# df['Td'] = df['Td'].astype(float)

df.tail(10)
```
